# Polynomial

```python
import torch
import numpy as np
```

```python
x_input = torch.tensor(5.0, requires_grad=True)
```

```python
def polynomial(x):
    return 3*x**2+2*x+1
```

```python
y= polynomial(x_input)
print(y)
```

```python
dy_dx=y.backward()
print(x_input.grad)
```

```python
print(x_input.grad)
```

# Vector valued function

```python
x_vector = torch.tensor([2.8,5.1], requires_grad=True)
print(x_vector)
```

```python
# Defining a non trivial function which ouputs a torch.tensor
def vector_function(x):
    y1 = x[0]**2
    y2 = x[1]**x[0]
    y3 = torch.sin(x[1])*x[1]
    return torch.stack([y1,y2,y3])


y = vector_function(x_vector)
print(y)
```

```python
# Creating the vector v which is of the same size as y
v=torch.tensor([1.0,1.0,1.0])

# perform the backward pass
y.backward(v)
```

```python
print(x_vector.grad)
```

# Differentiation of the parameters

```python
def function_params(x,params):
    return params[0]*torch.sin(x)+params[1]**3
```

```python
# Here I define parameters as torch.tensor which will be differentiated later on
params = torch.tensor([2.4,1.8], requires_grad=True)





# Let's see if the function is working
# Defining output which will be differentiated later
output1 = function_params(x_input, params)
print(output1)
```

```python
output1.backward()
```

```python
print(f"This is the output of params.grad: {params.grad}")
print(f"This is the ouput of x_input.grad: {x_input.grad}")
```

```python
# Similarly let me try the above where input is not differentaited
x_constant = torch.tensor(5.3, requires_grad=False)

output2 = function_params(x_constant, params)
# Now let me calculate the backward pass

output2.backward()


```
